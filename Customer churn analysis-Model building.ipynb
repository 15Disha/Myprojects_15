CUSTOMER CHURN PREDICTION (MODEL BUILDING PART)

```python
import pandas as pd
from sklearn import metrics
from imblearn.combine import SMOTEENN
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
df=pd.read_csv("D:\\telco_churn.csv")
```

```python
df.head()
```

```python
df=df.drop('Unnamed: 0',axis=1)
```

```python
df.head()
```

```python
#creating x variables

x=df.drop('Churn',axis=1)
x
```

```python
#Creating y variable 

y=df['Churn']
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

**ALGORITHMS**

Decision Tree Classifier

```python
model1=DecisionTreeClassifier()
```

```python
model1.fit(x_train,y_train)
```

```python
pred=model1.predict(x_test)
```

```python
print(classification_report(y_test,pred,labels=[0,1]))
```

```python
print(confusion_matrix(y_test,pred))
```

Random Forest Classifier

```python
model2=RandomForestClassifier()
```

```python
model2.fit(x_train,y_train)
```

```python
pred=model2.predict(x_test)
```

```python
pred
```

```python
print(classification_report(y_test,pred,labels=[0,1]))
```

```python
print(confusion_matrix(y_test,pred))
```

```python
samp=SMOTEENN()
x_resampled,y_resampled=samp.fit_resample(x,y)
```

```python
xr_train,xr_test,yr_train,yr_test=train_test_split(x_resampled,y_resampled,test_size=0.2)
```

Decision Tree Classifier

```python
model1_smote=DecisionTreeClassifier()
```

```python
model1_smote.fit(xr_train,yr_train)
```

```python
pred1_smote=model1_smote.predict(xr_test)
```

```python
print(classification_report(yr_test,pred1_smote,labels=[0,1]))
```

```python
print(confusion_matrix(yr_test,pred1_smote))
```

```python
model1_smote.score(xr_test,yr_test)
```

Random Forest Classifier

```python
model2_smote=RandomForestClassifier()
```

```python
model2_smote.fit(xr_train,yr_train)
```

```python
pred2_smote=model2_smote.predict(xr_test)
```

```python
print(classification_report(yr_test,pred2_smote,labels=[0,1]))
```

```python
print(confusion_matrix(yr_test,pred2_smote))
```

```python
model2_smote.score(xr_test,yr_test)
```
